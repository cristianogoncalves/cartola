```python
#python3
import pandas as pd
import numpy as np
import re
```

```python
planilhas = pd.read_excel(r'.\Excel\1rodada.xlsx')
planilhas['rodada'] = 1
for i in range(2,9):
    x = pd.read_pickle(r'.\Isoladas\{}rodada_isolada.pkl'.format(i))
    x['rodada'] = i
    planilhas = pd.concat([planilhas,x],axis=0)
```

```python
resultados = pd.read_excel(r'.\Excel\resultados.xlsx')
resultados['nrodada'] = resultados.rodada.apply(lambda x: int(re.findall('\d+',x)[0]))
time = resultados.Mandante.unique()

def casa_fora(df, time):
    return len(df[df['Mandante'] == time])

relacao_jogos = {}
for i in range(1,9):
    frame = resultados[resultados.nrodada == i]
    
    relacao_jogos[i] = {t : casa_fora(frame,t) for t in time}
```

```python
planilhas.drop(['Data','Ranking','Ranking_Visitante','adversario','hora','local','local2'],axis=1,inplace=True)

# Colocando o local correto da partida
planilhas['Local'] = planilhas.apply(lambda x: relacao_jogos[x.rodada][x.Time],axis=1)
planilhas['Local'] = planilhas['Local'].map({1: 'Casa',0: 'Fora'})
```

```python
planilhas['chave'] = planilhas.apply(lambda x: '{}-{}-{}'.format(x['Time'],x['Apelido'],x['Posicao']),axis=1)
```

```python
jogados = planilhas[planilhas['Atleta_pontos'] != 0]
```

```python
pontuacao = jogados.groupby(['chave','Local'])['Atleta_pontos'].agg({'media': np.mean,'desvio': np.std,'cont': np.count_nonzero})
```

```python
pontuacao = pontuacao.unstack('Local')
pontuacao = pontuacao.sort_values([('media','Casa')], ascending=False)
pontuacao.reset_index(inplace=True)
```

```python
pontuacao
```

```python
p = planilhas[planilhas.rodada == 8]
```

```python
nova = pd.merge(pontuacao, p[['Time','Abreviacao','Nome','Apelido','Posicao','chave']],how='left',on='chave')
```
