```python
import pandas as pd
import numpy as np
%pylab inline
```

```python
resultados = pd.read_excel(r'.\Excel\resultados.xlsx')
```

```python
def who_won(score_Mandante, score_Visitante):
    if score_Mandante > score_Visitante:
        return 'Mandante'
    elif score_Mandante < score_Visitante:
        return 'Visitante'
    else:
        return 'Empate'

resultados['resultado'] = resultados.apply(lambda x: who_won(x['Mandante_Score'],x['Visitante_Score']),axis=1)
```

```python
resultados.head()
```

```python
mandante = resultados.groupby(['Mandante','resultado']).count()['rodada']
mandante = mandante.unstack('resultado')
mandante.columns = ['Empatou','Ganhou','Perdeu']
```

```python
visitante = resultados.groupby(['Visitante','resultado']).count()['rodada']
visitante = visitante.unstack('resultado')
visitante.columns = ['Empatou','Perdeu','Ganhou']
visitante
```
